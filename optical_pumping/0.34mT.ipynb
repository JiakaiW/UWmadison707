```python

import csv

x = []
y = []
with open(name, newline='\n') as csvfile:
    reader = csv.DictReader(csvfile)
    for i,row in enumerate(reader):
        if i >= 14:
            x.append(row['Model']) 
            y.append(row['TBS2104'])
```

```python
import numpy as np

x = np.array(x,dtype=float)
y = np.array(y,dtype=float)
```

```python
import matplotlib.pyplot as plt
fig,axes = plt.subplots(2,1)

axes[0].plot(x)
axes[1].plot(y)
```

```python
fig,axes = plt.subplots(2,1)

axes[0].plot(y)
axes[0].set_xlim(280,380)

axes[1].plot(y)
axes[1].set_xlim(1600,1700)
```

```python
start_idx = 340
stop_idx = 1650
x_trunc = x[start_idx:stop_idx]
y_trunc = y[start_idx:stop_idx]
```

```python
stop_idx - start_idx
```

```python
freq_start = 20
freq_stop = 1002
```

```python
freq = np.arange(len(x_trunc))
freq = freq * (freq_stop-freq_start)/len(freq) + freq_start

plt.plot(freq)
plt.xlabel('idx')
plt.ylabel('KHz')
```

```python
plt.plot(freq,y_trunc)
plt.title('Measured 0.34 mT, +0.5 V ')
plt.xlabel('KHz')
plt.ylabel('response')
```
